```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Median_income'
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
print(soup)
```

```python
table = soup.find_all('table')[1]
world_titles = table.find_all('th')
print(world_titles)
```

```python
world_table = [title.text.strip() for title in world_titles]
print(world_table)
```

```python
print(world_table)
```

```python
import pandas as pd
```

```python
head = pd.DataFrame(columns = world_table)
head
```

```python
column_data = table.find_all('tr')
column_data
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    r_data = [data.text.strip() for data in row_data]
    length = len(head)
    head.loc[length] = r_data
    
```
